```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
```

```python
data = {'Message': [
    "Congratulations! You won a free vacation. Click here to claim.",
    "Hey, are we still meeting for lunch today?",
    "Limited time offer! Buy now and get 50% off!",
    "Can you send me the project report?",
    "URGENT: Your bank account is at risk. Verify now."
],
'Label': ['Spam', 'Ham', 'Spam', 'Ham', 'Spam']}
```

```python
df = pd.DataFrame(data)
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['Message'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, df['Label'], test_size=0.2, random_state=42)
model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
new_email = ["Win a brand new iPhone! Click the link below."]
new_email_vectorized = vectorizer.transform(new_email)
prediction = model.predict(new_email_vectorized)
```

```python
print("Prediction:", prediction[0])
```
